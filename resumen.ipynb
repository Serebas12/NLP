```python
import pandas as pd 

resumen= pd.read_csv("C:/Users/jssaa/OneDrive/Documentos/sebastian saavedra/Maestria/Semestre 2024 02/NLP/Proyecto/resumen_pruebas.csv",delimiter=";", encoding='latin1')

coord_persona=(4.628363, -74.064807)

resumen.head()
```

```python
resumen['latitude'] = resumen['latitude'].str.replace(',', '.').astype(float)
resumen['longitude'] = resumen['longitude'].str.replace(',', '.').astype(float)
resumen.head()
```

```python
from geopy.distance import geodesic

def calculate_distance(row, person_coords):
    restaurant_coords = (row['latitude'], row['longitude'])
    distance = geodesic(restaurant_coords, person_coords).kilometers  # devuelve la distancia en kilómetros
    return distance


resumen['distancia'] = resumen.apply(calculate_distance, args=(coord_persona,), axis=1)
resumen
```

```python
resumen.sort_values(by='distancia', ascending=True)
```
